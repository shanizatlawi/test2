```python
#6
import pandas as pd

# Load the data from the CSV file
items_df = pd.read_csv('C:/Users/97254/TEST_CSV/items.csv')

# 1. Add a new column that counts how many non-null values exist in each row
items_df['null_not_count'] = items_df.notnull().sum(axis=1)

# 2. Sort the DataFrame by item name and the new column (in descending order for null_not_count)
items_df = items_df.sort_values(by=['item_name', 'null_not_count'], ascending=[True, False])

# 3. Remove duplicates based on the item name (keep only the first occurrence)
items_df = items_df.drop_duplicates(subset='item_name', keep='first')

# Remove the temporary column if it is no longer needed
items_df = items_df.drop(columns=['null_not_count'])

# Display the result
print(items_df)

```

```python
#7
import pandas as pd

# Load the data
items_df = pd.read_csv('C:/Users/97254/TEST_CSV/items.csv')
order_item_df = pd.read_csv('C:/Users/97254/TEST_CSV/order_item.csv')

# Add a column for non-null counts
items_df['null_not_count'] = items_df.notnull().sum(axis=1)

# Sort and remove duplicates, keeping the one with the highest null_not_count
items_df = items_df.sort_values(by=['item_name', 'null_not_count'], ascending=[True, False])
duplicates_to_keep = items_df.drop_duplicates(subset='item_name', keep='first')

# Create a mapping of duplicate item IDs (from 'id' column in items_df) to the IDs to keep (also 'id' column)
# Correct the boolean mask alignment
mask = items_df.duplicated(subset='item_name', keep=False)
filtered_items_df = items_df[mask]

# Now create the mapping
mapping = filtered_items_df.set_index('id').reset_index().merge(
    duplicates_to_keep[['item_name', 'id']], on='item_name', suffixes=('_duplicate', '_keep')
)[['id_duplicate', 'id_keep']].set_index('id_duplicate')['id_keep'].to_dict()

# Update the item IDs in the order_item_df (where 'item_id' is used)
order_item_df['item_id'] = order_item_df['item_id'].replace(mapping)

# Save the updated order_item_df or display it
print(order_item_df)

# Optionally save the updated DataFrame to a new CSV file
order_item_df.to_csv('C:/Users/97254/TEST_CSV/updated_order_item.csv', index=False)
```
